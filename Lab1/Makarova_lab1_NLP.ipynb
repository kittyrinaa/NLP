### Task1
Using `FreqDist`, find the following from text5:
* words that are hapaxes (unique) and those that occur less than 5 times
* draw a plot showing mapping from word length to word frequency
* find 10 most frequent words occurring at the end of the sentence

```python
from nltk.probability import FreqDist
from nltk.book import *
import re
```

```python
fdist = FreqDist(text3)
print(fdist)
```

```python
sorted_dict = dict(sorted(fdist.items(), key=lambda item: item[1], reverse=True))
print("-"*10 + "Words that occur less than 5 times" + "-"*10)
for key, value in sorted_dict.items():
    if value < 5 and bool(re.match('^[a-zA-Z]+$', key)):
        print(f"{key}: {value} times")
```

```python
print("-"*10 + "Unique words" + "-"*10)
for key, value in sorted_dict.items():
    if value == 1 and bool(re.match('^[a-zA-Z]+$', key)):
        print(f"{key}: {value} times")
```

```python
from matplotlib import pyplot as plt
freq = {}
for key, value in sorted_dict.items():
    if bool(re.match('^[a-zA-Z]+$', key)):
        if len(key) not in freq:
            freq[len(key)] = value
        else:
            freq[len(key)] += value
            
freq = dict(sorted(freq.items(), key=lambda item: item[0], reverse=True))
plt.plot(freq.keys(), freq.values(), marker = '*',  color = 'darkblue') 
plt.title("Mapping from word length to word frequency")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()
```

```python
from collections import Counter
from nltk.tokenize import sent_tokenize, word_tokenize

sentences = sent_tokenize(" ".join(text3))
end_freq = Counter()

for sentence in sentences:
    words = word_tokenize(sentence)
    for word in words[::-1]:
        if bool(re.match('^[a-zA-Z]+$', word)):
            end_freq[word] += 1
            break
            
sorted_freq = sorted(end_freq, key=lambda x: end_freq[x], reverse=True)
print("-"*10 + "Ten most frequen words at the end of a sentence" + "-"*10)
for word in sorted_freq[:10]:
    print(f"{word}: {end_freq[word]} times")
```

### Task2

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
with open("Collins Suzanne. The hunger games.txt") as f:
    raw = f.read()
    tokens = nltk.word_tokenize(raw)
    text = nltk.Text(tokens)
```

```python
text.concordance("Katniss")
```

```python
text.similar("fire")
```

```python
from nltk.draw.dispersion import dispersion_plot
dispersion_plot(text, ["Katniss", "Peeta", "death"])
```

```python
fdist = FreqDist(text)
print(fdist)
```

```python
fdist.most_common(50)
```
